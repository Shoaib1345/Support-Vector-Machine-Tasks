# <font color="purple"><h3 align="center"> Implement SVM using hyperparameter Tuning on Hotel_Bookings_Clean</h3></font>

```python
import pandas as pd
import numpy as np
from sklearn.matrics import accuray_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
```

```python
hostel=pd.read_csv('hotel_bookings_clean.csv')
hostel
```

```python
hostel.shape
```

```python
hostel.head(10)
```

```python
hostel.isna().sum()
```

```python
hostel.info()
```

```python
X=hostel.iloc[:,1:]
Y=hostel['is_canceled']
X
```

```python
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.2, random_state=123)
hyperparams={'kernel':'linear',
       'C':1,
       'gamma':0.01,
       'degree':3,
      'random_state':123}
```

```python
rs_param_grid={'kernel':['linear','poly','rbf','sigmoid'],
       'C':[0.1,1,10],
       'gamma':[0.00001,0.0001,0.001,0.01,0.1]
       }
```

```python
from sklearn.svm import SVC
clfs=SVC(**hyperparams)
clfs.fit(X_train,Y_train)
```

```python
test_prediction=clfs.predict(x_test)
```

```python
acc=accuracy_score(test_prediction,y_test)
acc
```

```python
clf=SVC(random_state=123)
clf_rs=RandomizedSearchCV(estimator=clf,
                         param_distributions=rs_param_grid,
                         cv=3,
                         n_iter=5,
                         verbose=2,
                         random_state=123)
```

```python
clf_rs.fit(X_train,Y_train)
```

```python
print(clf_rs.best_params_)
print(clf_rs.best_score_)
```

# <font color="purple"><h3 align="center">Implement SVM on Iris Dataset</h3></font>

```python
#task 02
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
X=iris.data
Y=iris.target
```

```python
iris_pd=pd.DataFrame(data=X, columns=iris.feature_names)
iris_pd['target']=Y
iris_pd
```

```python
X=iris_pd.drop('target',axis=1)
Y=iris_pd['target']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=43)
X_test
```

```python
from sklearn.svm import SVC
svc_model=SVC(kernel='linear', C=1.0)
svc_model.fit(X_train,Y_train)
```

```python
test_prediction=svc_model.predict(X_test)
train_pre=svc_model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(test_prediction,Y_test)
acc_train=accuracy_score(train_pre,Y_train)
print("Accuracy: {:.2f}%".format(acc))
print("Accuracy: {:.2f}%".format(acc_train))
```
